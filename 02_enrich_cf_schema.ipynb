```python
import csv
import json
```

```python
# Add general.csv
json_object = None
with open('rules-generation/general.csv', newline='') as csvfile:
    reader = csv.DictReader(csvfile)
    # Opening CF resource JSON file
    f = open('rules/cf_resource_provider_schema.json')
    json_object = json.load(f)
    for row in reader:
        type_name = row['type name CF'].rstrip("\n")
        property_name = row['cf_property_desc'].rstrip("\n")

        if type_name == "-":
            type_name = ""
        if property_name == "-":
            property_name = ""

        # If false insert nothing    
        if type_name and property_name:
            if not type_name in json_object:
                json_object[type_name] = {}
            if not property_name in json_object[type_name]:
                json_object[type_name][property_name] = {}

            dict_property = json_object[type_name][property_name]
            dict_to_be_added = {}

            if row['Our own regex']:
                dict_to_be_added["pattern"] = row['Our own regex'].rstrip("\n")
            if row['Rule in plain English']:
                dict_to_be_added["explanation"] = row['Rule in plain English'].rstrip("\n")
            if row['Alternative regex']:
                dict_to_be_added["alt_pattern"] = row['Alternative regex'].rstrip("\n")
            if row['can be deleted?']:
                dict_to_be_added["delete_possible"] = row['can be deleted?'].rstrip("\n")
            
            dict_property["name_it_right"] = dict_to_be_added
            json_object[type_name][property_name] = dict_property

# Save to file
json_object_new = json.dumps(json_object, indent=4)
with open("rules/rules_merged.json", "w") as outfile:
    outfile.write(json_object_new)
```

```python
# Add granular errors
json_object = None
with open('rules-generation/errors.csv', newline='') as csvfile:
    reader = csv.DictReader(csvfile)

    f = open('rules/rules_merged.json')
    json_object = json.load(f)

    for row in reader:
        type_name = row['Service type_name'].rstrip("\n")
        property_name = row['property'].rstrip("\n")
        error_id = row['error_id'].rstrip("\n")
        regex = row['regex'].rstrip("\n")
        error_message = row['error_message'].rstrip("\n")

        # If false insert nothing    
        if type_name and property_name and error_id and regex and error_message:
            if not type_name in json_object:
                json_object[type_name] = {}
            if not property_name in json_object[type_name]:
                json_object[type_name][property_name] = {}
            if not "name_it_right" in json_object[type_name][property_name]:
                json_object[type_name][property_name]["name_it_right"] = {}
            if not "errors" in json_object[type_name][property_name]["name_it_right"]:
                json_object[type_name][property_name]["name_it_right"]["errors"] = []

            dict_temp = {
                "id" : error_id,
                "regex" : regex,
                "errorMessage" : error_message,
            }

            json_object[type_name][property_name]["name_it_right"]["errors"].append(dict_temp)

# Save to file
json_object_new = json.dumps(json_object, indent=4)
with open("rules/rules_merged.json", "w") as outfile:
    outfile.write(json_object_new)
```
